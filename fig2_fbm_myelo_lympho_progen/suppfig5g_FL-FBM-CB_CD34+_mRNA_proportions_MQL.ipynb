```python
#SETTING ENVIRONMENT 
import rpy2
#import tzlocal
#%load_ext rpy2.ipython

import numpy as np
import pandas as pd
import scanpy as sc

sc.settings.verbosity = 3             # verbosity: errors (0), warnings (1), info (2), hints (3)
#sc.logging.print_versions()
sc.settings.set_figure_params(dpi=80, dpi_save=1000)
```

```python
adata = sc.read("h5da_part4/050221_mq224_mRNA_Progenitors_only_postHarmony_ccfiltered_DEG_P4_sorted_annotations_FBMS2-3-H3-F3-E5_CB-G7_FL-A7-A6-C5.h5ad")
```

```python
#FL 
adata1 = adata[adata.obs["Tissue"]=="FL",:].copy()
#FBM
adata2= adata[adata.obs["Tissue"]=="FBM",:].copy()
#CB
adata3 = adata[adata.obs["Tissue"]=="CB",:].copy()
```

```python
tab1 = pd.crosstab(adata1.obs["assignment"], adata1.obs['Cell.labels.P4.sorted'], normalize="index")*100
tab2 = pd.crosstab(adata2.obs["assignment"], adata2.obs['Cell.labels.P4.sorted'], normalize="index")*100
tab3 = pd.crosstab(adata3.obs["assignment"], adata3.obs['Cell.labels.P4.sorted'], normalize="index")*100
```

```python
tab2
```

```python
tab3
```

```python
average1 = np.mean(tab1, axis=0)
average2 = np.mean(tab2, axis=0)
average3 = np.mean(tab3, axis=0)
std1 = np.std(tab1, axis=0)
std2 = np.std(tab2, axis=0)
std3 = np.std(tab3, axis=0)
df = pd.concat([average1,average2,average3,std1, std2,std3],axis=1, join='inner')
df.columns = ["Mean FL","Mean FBM","Mean CB","SD FL","SD FBM","SD CB"]
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df[["Mean FL","Mean FBM","Mean CB"]].plot(kind='bar', yerr=df[["SD FL","SD FBM","SD CB"]].values.T, alpha = 0.5, error_kw=dict(ecolor='k'),width=0.8)
plt.rcParams["figure.figsize"] = (10.8,10.8)
plt.legend(loc=(0.2,1.01))
plt.grid(False)
plt.savefig('./Proportions_part4.4/050221_mq224_mRNA_Progenitors_only_ccfiltered_DEG_P4_sorted_annotations_FBMS2-3-H3-F3-E5_CB-G7_FL-A7-A6-C5.pdf')
plt.show()
```
